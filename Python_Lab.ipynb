```python
# Put these commands in the first cell of your notebook, and run immediately

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading of  all files
sales = pd.read_csv('./sales.csv')
sales_rep = pd.read_csv('./sales_rep.csv')
items = pd.read_csv('./item_master.csv')
customers = pd.read_csv('./customers.csv', encoding='latin-1')
cities = pd.read_csv('./cities.csv', encoding='latin-1')
```

```python
# Merge into one big 'sales' frame
cust = customers.merge(cities,how='left',on='City Code')
sales = sales.merge(sales_rep,how='inner',left_on='Sales Rep Number', right_on='Sales Rep ID')
sales = sales.merge(items,how='inner',on='Item Number')
sales = sales.merge(cust,how='inner',on='Customer Number')
```

```python
# This graph below was used to see how orders grew or shrank over time. 
#It is not the best chart but it did inspire the next two charts!


sales['Invoice Date'] = pd.to_datetime(sales['Invoice Date'])
sales.sort_values(by='Invoice Date')

sales['Invoice Date'].value_counts().sort_index().plot()
```

```python
#Sales by quarter. 2014 is off to a slow start. There has been a seasonal lift in sales count going from the 2Q to the 3Q.

sales['Quarter_Info'] = sales['Invoice Date'].dt.to_period('Q')


sales['Quarter_Info'] = (sales['Quarter_Info'])
sales.sort_values(by='Quarter_Info')

sales['Quarter_Info'].value_counts().sort_index().plot()

plt.xlabel("Quarterly Dates", size=12)
plt.ylabel("Count of Sales", size=12)
plt.title("Sales by Quarter", size=14);
```

```python
# The formula above was recycled and adjusted to monthy.  We also see a decline in sales going into April. 

sales['Quarter_Info'] = sales['Invoice Date'].dt.to_period('M')


sales['Quarter_Info'] = (sales['Quarter_Info'])
sales.sort_values(by='Quarter_Info')

sales['Quarter_Info'].value_counts().sort_index().plot()

plt.xlabel("Quarterly Month", size=12)
plt.ylabel("Count of Sales", size=12)
plt.title("Sales by Month", size=14);
```
